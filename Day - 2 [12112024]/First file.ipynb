```python
#Assignment
a = 10
b = 20
print(a)
print(b)
#Reassignment
a = 100
a = 300
print(a)
#Multiple assignment
a,b,c = 10,20,30
print(a)
print(b)
print(c)
a=b=c=34
print(a)
print(b)
print(c)
#Swapping of varaible
a = 10
b = 20
print(a)
print(b)
a,b = b,a
print(a)
print(b)
#Deleting Variable
hello = 123
print(hello)
del hello
print(hello)
```

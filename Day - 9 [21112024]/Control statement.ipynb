#While Loop

```python
#Printing elements from 1 to 10
#Approach - 1
'''
n = 0
while n<10:
  print(n+1)
  n = n + 1
'''
#Appraoch - 2
'''
n = 1
while n<=10:
  print(n)
  n = n + 1
'''
```

#For Loop

```python
#Printing numbers from 1 to 10
#Approach - 1
'''
for i in range(10):
  print(i+1)
'''
#Approach - 2
'''
for i in range(1,11):
  print(i)
'''
```

#Break

```python
#Break with for loop
'''
for i in range(1,11):
  if  i == 5:
    break
  else:
    print(i)
'''
#Break with while loop
'''
i = 1
while i <= 10:
  if i==5:
    break
  else:
    print(i)
  i = i + 1
'''
```

#Continue

```python
#Continue with for loop
'''
for i in range(1,11):
  if i == 5:
    print("Condition mathced")
    continue
    print("continue executed")
  else:
    print(i)
'''
#Continute with while loop
'''
i = 1
while i<=10:
  if i==5:
    i = i + 1
    print("Condition mathced")
    continue
    print("continue executed")
  else:
    print(i)
  i = i + 1
'''
```

#Pass

```python
for i in range(10):
  pass
```

#While With else

```python
i = 1
while i<=10:
  if i == 5:
    pass
  else:
    print(i)
  i = i + 1
else:
  print("Else executed")
```

#For with else

```python
for i in range(1,11):
  if  i == 5:
    pass
  else:
    print(i)
else:
  print("else Executed")
```
